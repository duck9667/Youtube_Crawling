```python
import re
```

```python
import pandas as pd
df = pd.read_excel("crawling.xlsx")
```

```python
df.columns
```

```python
comments = df[['name','comment_list']]
```

```python
comments
```

```python
pre_comment = []
for i in range(5) :
    pre_comment.append("".join(comments['comment_list'][i]))
```

```python
pre_comment
```

```python
def clean(text) :
    emoji_pattern = re.compile("["u"\U0001F600-\U0001F64F"  # emoticons
                                  u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                                  "]+", flags=re.UNICODE)
    
    clean_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]','', str(text))
    cleaned_text = re.sub(emoji_pattern,'',clean_text)
    return cleaned_text
```

```python
clean_comment = clean(pre_comment)

pattern = re.compile(r'[^가-힣]') 
clean_comment = re.sub(pattern,' ', clean_comment)
clean_comment
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
komoran.pos(clean_comment)
```

```python
from konlpy.tag import Okt
okt = Okt()
okt.pos(clean_comment)
```
